```python
# Carregar a planilha e depois selecionar as colunas de interesse
df_notas = pd.read_parquet(r"C:\Users\hugob\Downloads\df_EF_EM_bncc_censo.parquet")
```

```python
df_notas
```

```python
df['CPF'] = df['CPF'].astype(str).str.replace('.', '', regex=False).str.replace('-', '', regex=False)
```

```python
# MErge para adicionar idade no dataframe df_notas

df_notas = df_notas.merge(
    df[['CPF', 'IDADE']],      # DataFrame com as colunas que você quer trazer
    how='left',                # mantém todas as linhas do df_notas
    left_on='CPF PESSOA',      # coluna de df_notas usada para combinar
    right_on='CPF'             # coluna de df usada para combinar
)
```

```python
df_notas
```

```python
# Função auxiliar para pegar o valor mais frequente (moda)
def moda(serie):
    return serie.mode()[0] if not serie.mode().empty else None

# Agrupar e calcular
df_reprovacoes = (
    df_notas.groupby('CPF PESSOA', as_index=False)
    .agg({
        'STATUS': lambda x: (x == 'Reprovado').sum(),  # Conta quantos "Reprovado"
        'NOME PESSOA': moda,
        'IDADE': moda,                                # Valor mais frequente da idade
        'SÉRIE': moda,                                 # Valor mais frequente da série
        'DIREC': moda,
        'MUNICÍPIO': moda,
        'ESCOLA': moda,
        'INEP ESCOLA': moda,
        'ETAPA ENSINO': moda,
    })
)

# Renomear a coluna resultante
df_reprovacoes.rename(columns={'STATUS': 'TOTAL_REPROVACOES'}, inplace=True)
```

```python
df_reprovacoes
```

```python
df_reprovacoes['SÉRIE'].unique()
```

```python
# Dicionário de mapeamento entre série e idade esperada
mapa_idade_esperada = {
    '6º ANO': 11,
    '7º ANO': 12,
    '8º ANO': 13,
    '9º ANO': 14,
    '1ª SÉRIE': 15,
    '2ª SÉRIE': 16,
    '3ª SÉRIE': 17
}

# Criar nova coluna com base na série
df_reprovacoes['IDADE_ESPERADA'] = df_reprovacoes['SÉRIE'].map(mapa_idade_esperada)
```

```python
df_reprovacoes
```

```python
# Diferença entre Idade e Idade Esperada
df_reprovacoes['DIFERENCA'] = df_reprovacoes['IDADE'] - df_reprovacoes['IDADE_ESPERADA']
```

```python
df_reprovacoes
```

```python
# Coluna de identificação dos estudantes em distorção idade-série (2 ou mais anos de diferença entre idade atual e idade esperada para a série)
df_reprovacoes['DISTORCAO'] = np.where(df_reprovacoes['DIFERENCA'] >= 2, 'Sim', 'Não')
```

```python
df_reprovacoes
```

```python
# Dataframe só com os estudantes com distção idade-série: 2 ou mais anos de diferença entre idade atual e idade esperada para a série)
df_distorcao = df_reprovacoes[df_reprovacoes['DISTORCAO'] == 'Sim'].copy()
```

```python
df_distorcao
```

```python
# COluna de Etapa
# Definir as condições
condicoes = [
    df_distorcao['SÉRIE'].isin(['6º ANO', '7º ANO', '8º ANO', '9º ANO']),
    df_distorcao['SÉRIE'].isin(['1ª SÉRIE', '2ª SÉRIE', '3ª SÉRIE'])
]

# Definir os valores correspondentes
valores = [
    'Ensino Fundamental - Anos Finais',
    'Ensino Médio'
]

# Criar a nova coluna 'ETAPA'
df_distorcao['ETAPA'] = np.select(condicoes, valores, default='Outro')
```

```python
df_distorcao
```

```python
# Garantir que TOTAL_REPROVACOES é numérico
df_distorcao['TOTAL_REPROVACOES'] = pd.to_numeric(df_distorcao['TOTAL_REPROVACOES'], errors='coerce')

# Calcular as quantidades em cada faixa
qtd_7_ou_mais = (df_distorcao['TOTAL_REPROVACOES'] >= 7).sum()
qtd_4_a_6 = df_distorcao['TOTAL_REPROVACOES'].between(4, 6).sum()
qtd_3_ou_menos = (df_distorcao['TOTAL_REPROVACOES'] <= 3).sum()

# Mostrar o resultado
print("Total em distorção idade-série:", len(df_distorcao))
print("7 ou mais reprovações:", qtd_7_ou_mais)
print("4, 5 ou 6 reprovações:", qtd_4_a_6)
print("3 ou menos reprovações:", qtd_3_ou_menos)
```

```python
# Cálculo por etapa:
# Agrupar por ETAPA e calcular as quantidades em cada faixa
resultado_por_etapa = (
    df_distorcao
    .groupby('ETAPA')
    .apply(lambda g: pd.Series({
        'Total em distorção idade-série': len(g),
        '7 ou mais reprovações': (g['TOTAL_REPROVACOES'] >= 7).sum(),
        '4, 5 ou 6 reprovações': g['TOTAL_REPROVACOES'].between(4, 6).sum(),
        '3 ou menos reprovações': (g['TOTAL_REPROVACOES'] <= 3).sum()
    }))
    .reset_index()
)
```

```python
resultado_por_etapa
```

```python
# Cálculo por SÉRIE:
# Agrupar por SÉRIE e calcular as quantidades em cada faixa
resultado_por_serie = (
    df_distorcao
    .groupby('SÉRIE')
    .apply(lambda g: pd.Series({
        'Total em distorção idade-série': len(g),
        '7 ou mais reprovações': (g['TOTAL_REPROVACOES'] >= 7).sum(),
        '4, 5 ou 6 reprovações': g['TOTAL_REPROVACOES'].between(4, 6).sum(),
        '3 ou menos reprovações': (g['TOTAL_REPROVACOES'] <= 3).sum()
    }))
    .reset_index()
)
```

```python
resultado_por_serie
```

```python
# Excluir coluna idade, pois não é útil aqui
df_reprovacoes = df_reprovacoes.drop(columns=['IDADE', 'ETAPA ENSINO'])
```

```python
# Coluna de Etapa
# Definir as condições
condicoes = [
    df_reprovacoes['SÉRIE'].isin(['6º ANO', '7º ANO', '8º ANO', '9º ANO']),
    df_reprovacoes['SÉRIE'].isin(['1ª SÉRIE', '2ª SÉRIE', '3ª SÉRIE'])
]

# Definir os valores correspondentes
valores = [
    'Ensino Fundamental - Anos Finais',
    'Ensino Médio'
]

# Criar a nova coluna 'ETAPA'
df_reprovacoes['ETAPA'] = np.select(condicoes, valores, default='Outro')
```

```python
# Reordenar as colunas:
ordem = ['DIREC',
         'MUNICÍPIO',
         'ESCOLA',
         'INEP ESCOLA',
         'ETAPA',
         'SÉRIE',
         'CPF PESSOA',
         'NOME PESSOA',
         'TOTAL_REPROVACOES']

df_reprovacoes = df_reprovacoes[ordem]
```

```python
# Fazer Dataframe só com os estudantes de Ens. Fundamental com 2 ou 3 reprovações e estudantes de Ens. Médio com 5 ou 6 reprovações

df_reprovacoes_risco = df_reprovacoes[
    ((df_reprovacoes['ETAPA'] == 'Ensino Fundamental - Anos Finais')
    & (df_reprovacoes['TOTAL_REPROVACOES'].isin([2, 3])))
    |
    ((df_reprovacoes['ETAPA'] == 'Ensino Médio')
    & (df_reprovacoes['TOTAL_REPROVACOES'].isin([5, 6])))
].copy()
```
